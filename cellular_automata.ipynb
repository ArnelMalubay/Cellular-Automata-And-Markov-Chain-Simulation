### Setting-up the Simulation

Importing libraries

```python
import agentpy as ap
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
np.set_printoptions(suppress = True)
import IPython
```

Loading initial states and setting parameters

```python
# In order to run agentpy.Experiments, all model parameters must be hashable, so numpy arrays are converted to byte sequences
initial_states_array = np.load('initial_states.npy')
initial_states = initial_states_array.tobytes()
height = initial_states_array.shape[0]
width = initial_states_array.shape[1]
num_pixels_within_territory = np.sum(initial_states_array != 4)
seed = 8
```

```python
print('--------------------------------------------------')
for i, c in enumerate(['Forests and Wooded Land', 'Open Land', 'Built-up Areas', 'Water']):
    print(c)
    pixel_count = np.sum(initial_states_array == i)
    print(f'Pixel Count = {pixel_count}')
    print(f'Percentage = {100 * (pixel_count / num_pixels_within_territory)}')
    print(f'Land Cover = {pixel_count * 0.00180418}')
    print('--------------------------------------------------')
```

Defining the model class

```python
class MapModel(ap.Model):

    # adds the agents and their corresponding initial states to a grid with the same shape as initial_states
    def setup(self):
        n_pixels = int(self.p.height * self.p.width)
        pixels = self.agents = ap.AgentList(self, n_pixels)
        initial_states_frombuffer = np.frombuffer(self.p.initial_states, dtype = np.uint8).reshape(self.p.height, self.p.width)
        self.map = ap.Grid(self, (self.p.height, self.p.width))
        self.map.add_agents(pixels, [(x, y) for x in range(self.p.height) for y in range(self.p.width)])
        # each agent has a previous and current state so that all state updates are based on previous states and not updated ones
        for x in range(self.p.height):
            for y in range(self.p.width):
                self.map.agents[x, y].prev_state = initial_states_frombuffer[x, y]
                self.map.agents[x, y].current_state = initial_states_frombuffer[x, y]
        self.transition_matrix_frombuffer = np.frombuffer(self.p.transition_matrix).reshape(4, 4)

    # for each agent/pixel, consider all neighbors within a 9x9 region and compute a weighted probability based on the transition matrix
    def step(self):
        for pixel in self.agents:
            # pixels assigned to outside territory will never change
            if pixel.prev_state == 4: 
                pixel.current_state = 4
            else:
                w_j = [0] * 4
                for neighbor in self.map.neighbors(pixel, distance = 3):
                    if neighbor.prev_state == 4:
                        continue
                    else:
                        w_j[neighbor.prev_state] += 1
                probs = [w_j[j] * self.transition_matrix_frombuffer[pixel.prev_state, j] for j in range(4)]
                # in case all the relative probabilities are 0, all states are equally likely to be chosen as next state
                if sum(probs) > 0:
                    next_state = random.choices([i for i in range(4)], probs, k = 1)[0]
                else:
                    next_state = random.choices([i for i in range(4)], [1] * 4, k = 1)[0]
                pixel.current_state = next_state
        # once all agents have been updated, update all previous states
        for pixel in self.agents:
            pixel.prev_state = pixel.current_state
    
    # records the proportion of land use states at each step
    def update(self):
        for i, c in enumerate(['Forests and Wooded Land', 'Open Land', 'Built-up Areas', 'Water']):
            self.record(c, len(self.agents.select(self.agents.current_state == i)) / self.p.num_pixels_within_territory)
    
    # reports the proportion of land use states at the end
    def end(self):
        for i, c in enumerate(['Forests and Wooded Land', 'Open Land', 'Built-up Areas', 'Water']):
            self.report(c, len(self.agents.select(self.agents.current_state == i)) / self.p.num_pixels_within_territory)
```

Visualization

```python
# visualizes the actual map and the changes for each timestep
def land_use_plot(model, ax, title = 'Land Change Simulation'):
    attr_grid = model.map.attr_grid('current_state')
    color_dict = {0 : '#00FF00', 1 : '#FFFF00', 2 : '#FF0000', 3 : '#0000FF', 4 : '#FFFFFF'}
    ax.set_title(title)
    ax.xaxis.set_tick_params(labelbottom = False)
    ax.yaxis.set_tick_params(labelleft = False)
    ax.set_xticks([])
    ax.set_yticks([])
    ap.gridplot(attr_grid, ax = ax, color_dict = color_dict, convert = True)

# visualizes the change in land use proportion
def land_stackplot(data, ax):
    x = data.index.get_level_values('t')
    y = [data[var] for var in ['Water', 'Built-up Areas', 'Open Land', 'Forests and Wooded Land']]
    sns.set()
    ax.stackplot(x, y, labels = ['Water', 'Built-up Areas', 'Open Land', 'Forests and Wooded Land'], colors = ['#0000FF', '#FF0000', '#FFFF00', '#00FF00'])
    ax.legend()
    ax.set_xlim(0, max(1, len(x) - 1))
    ax.set_ylim(0, 1)
    ax.set_title('Land Use Percentage')
    ax.set_xlabel('Time steps')
    ax.set_ylabel('Percentage of Land')

# functions to be used for animation
def animation_plot(model, axs):
    ax1, ax2 = axs
    land_use_plot(model, ax1)
    land_stackplot(model.output.variables.MapModel, ax2)

def save_animation(model, run, steps = 50):
    fig, axs = plt.subplots(1, 2, figsize = (20, 7.5), width_ratios = [2, 1])
    animation = ap.animate(model, fig, axs, animation_plot, steps)
    animation.save(f'Results/run{run}/simulation.gif')
```

### Transition Matrix 1 - Mactan Island

```python
transition_matrix1 = np.load('Transitions/run1.npy').tobytes()

parameters1 = {'height' : height,
               'width' : width,
               'num_pixels_within_territory' : num_pixels_within_territory,
               'initial_states' : initial_states,
               'seed' : seed,
               'transition_matrix' : transition_matrix1}

model1 = MapModel(parameters1)
```

```python
np.frombuffer(transition_matrix1).reshape(4,4)
```

Visualizing one run

```python
save_animation(model1, 1)
```

```python
IPython.display.HTML('<img src="Results/run1/simulation.gif">')
```

Viewing Results of one run

```python
sim_df1 = model1.output.variables.MapModel
sim_df1.to_csv('Results/run1/sim_results.csv')
sim_df1
```

```python
sim_df1.plot.line(color = {'Forests and Wooded Land' : '#00FF00', 'Open Land' : '#FFFF00', 'Built-up Areas' : '#FF0000', 'Water' : '#0000FF'})
plt.savefig('Results/run1/lineplot.png')
```

```python
fig, ax = plt.subplots()
land_stackplot(sim_df1, ax)
plt.savefig('Results/run1/stackplot.png')
```

Visualizing the final configuration

```python
fig, ax = plt.subplots(figsize = (10, 10))
land_use_plot(model1, ax, '')
plt.savefig('Results/run1/final_config.png')
```

Running multiple runs

```python
parameters1['steps'] = 50
exp1 = ap.Experiment(MapModel, parameters1, iterations = 5)
results1 = exp1.run(n_jobs = 5)
```

```python
mult_sim_df1 = results1.reporters
mult_sim_df1.to_csv('Results/run1/mult_sim_results.csv')
mult_sim_df1
```

```python
def plot_hist(results, run):
    fig, ax = plt.subplots(2, 2, figsize = (15, 12), sharex = False, sharey = False)
    for i in range(4):
        sns.histplot(data = results.reporters, x = results.reporters[results.reporters.columns[i + 1]], ax = ax[i // 2, i % 2], kde = True)
    plt.savefig(f'Results/run{run}/mult_sim_comparison.png')
```

```python
plot_hist(results1, 1)
```
